```python
import pandas as pd
import numpy as np
```

```python
EXCEL_OPNION_PATH=r'Person1.xlsx'
```

```python
df=pd.read_excel(EXCEL_OPNION_PATH,header=None, index_col=None)
df
```

```python
# Convert the dataframe to a numpy array
matrix = df.values

# Step 2: Calculate the sum of each column
column_sums = np.sum(matrix, axis=0)

# Step 3: Normalize the matrix
normalized_matrix = matrix / column_sums

# Step 4: Calculate the priority vector (average of each row in the normalized matrix)
priority_vector = np.mean(normalized_matrix, axis=1)

# Step 5: Calculate the consistency vector (multiply the original matrix by the priority vector)
consistency_vector = np.dot(matrix, priority_vector)

# Step 6: Compute the principal eigenvalue (λ_max)
lambda_max = np.mean(consistency_vector / priority_vector)

# Step 7: Calculate the Consistency Index (CI)
n = matrix.shape[0]  # Number of criteria
ci = (lambda_max - n) / (n - 1)

# Step 8: Get the Random Index (RI) value for n criteria
# RI values can be found in standard AHP tables. Here's a small sample:
ri_table = {1: 0, 2: 0, 3: 0.58, 4: 0.90, 5: 1.12, 6: 1.24, 7: 1.32, 8: 1.41, 9: 1.45, 10: 1.49}
ri = ri_table.get(n, 1.49)  # If n > 10, use the value for n = 10

# Step 9: Calculate the Consistency Ratio (CR)
cr = ci / ri

# Output the results
print(f"Principal Eigenvalue (λ_max): {lambda_max}")
print(f"Consistency Index (CI): {ci}")
print(f"Random Index (RI): {ri}")
print(f"Consistency Ratio (CR): {cr}")

# Checking for acceptable consistency
if cr < 0.1:
    print("The consistency ratio is acceptable.")
else:
    print("The consistency ratio is not acceptable. Consider revising the judgments.")
```
